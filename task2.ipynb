```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("Titanic_Cleaned.csv")

# Display first 5 rows
print(df.head())

# Dataset information
print(df.info())

# Descriptive statistics
print(df.describe())

# Correlation
print(df.corr(numeric_only=True))

# Histograms
df.hist(figsize=(10,8))
plt.show()

# Boxplot
df.boxplot(figsize=(10,6))
plt.show()

# Summary
print("Average Age:", df["Age"].mean())
print("Average Fare:", df["Fare"].mean())
print("Survival Count:")
print(df["Survived"].value_counts())
print("Passenger Class Count:")
print(df["Pclass"].value_counts())
```

# Exploratory Data Analysis (EDA) - Titanic Dataset

## Findings

- The dataset contains 887 rows and 8 columns.
- No duplicate records were found.
- Missing values were checked before analysis.
- The average age of passengers is approximately 29.47 years.
- The average fare paid is approximately 32.31.

- The Fare column contains some outliers.
